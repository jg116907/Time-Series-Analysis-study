# 백색 잡음
- 시계열 분석을 구성하는 여러가지 기본 모형 중 가장 중요한 것
- 개별 확률 변수들이 서로 독립이고 동일한 확률 분포를 따르는 확률 과정 => i.i.d. 가정
> 1. 정상 과정이다.
  2. 시차(lag)가 0일 경우, 자기공분산은 확률 분포의 분산이 되고 시차가 0이 아닌 경우, 자기 공분산은 0이다.
  3. 시차(lag)가 0일 경우, 자기상관계수는 1이 되고 시차가 0이 아닌 경우, 자기상관계수는 0이다.

```
import scipy.stats as sp
import matplotlib.pyplot as plt
import numpy as np
```

```
# 가우시안 백색 잡음 - 확률 분포가 표준 가우시안 정규 분포
e = sp.norm.rvs(size=300)
plt.plot(e)
plt.show()
```

```
# 비-가우시안 백색 잡음 - 정규분포가 아닐 경우
e = sp.bernoulli.rvs(0.5,size=100)*2-1 # 베르누이 분포
plt.step(np.arange(len(e)),e)
plt.ylim(-1.1,1.1)
plt.show()
```

# 이산 시간 랜덤 워크(discrete-time random walk)
- 백색 잡음을 누적한 확률 과정

```
for i in range(3):
  np.random.seed(9*i)
  e = sp.norm.rvs(size=900)
  W = np.insert(np.cumsum(e),0,0)
  plt.plot(W)
plt.show()
```
